Definition:

1   **H₀:**  μ=x,  **H₁:**  μ≠x
2   **H₀:**  μ≤x,  **H₁:**  μ>x
3   **H₀:**  μ≥x,  **H₁:**  μ<x


1.- **H₀:** Null hypothesis is: the sample mean is equal to population mean (or... mu is equal to X).
2.- **H₀:** Null hypothesis is: the sample mean is less-than or equal to population mean (or... mu is less-than or equal to X).
3.- **H₀:** Null hypothesis is: the sample mean is greater-than or equal to population mean (or... mu is greater-than or equal to X).

To decide whether to use the parametric or nonparametric version of the test, we should check the specific requirements listed below:

Observations in each sample are independent and identically distributed (IID).
Observations in each sample are normally distributed.
Observations in each sample have the same variance.

Selecting the proper test

```python
# Generar dos conjuntos de datos aleatorios
np.random.seed(42)  # Para reproducibilidad
data1 = np.random.normal(loc=10, scale=2, size=1000)  # Media=10, Desviación estándar=2
data2 = np.random.normal(loc=12, scale=2, size=1000)  # Media=12, Desviación estándar=2

# Realizar la prueba de hipótesis (Prueba t de Student)
t_statistic, p_value = stats.ttest_ind(data1, data2)

alpha = 0.05  # Nivel de significancia

# Comparar p-value con el nivel de significancia
if p_value < alpha:
    print("Rechazar la hipótesis nula")
else:
    print("Aceptar la hipótesis nula")
    
print("T-statistic:", t_statistic)
print("P-value:", p_value)
```

```python
mean, std_dev = stats.norm.fit(data1)
data1Linspace = np.linspace(min(data1), max(data1), 100)
data1pdf = stats.norm.pdf(data1Linspace, mean, std_dev)

mean, std_dev = stats.norm.fit(data2)
data2Linspace = np.linspace(min(data1), max(data2), 100)
data2pdf = stats.norm.pdf(data2Linspace, mean, std_dev)

# Grafica la PDF
plt.figure(figsize=(8, 6))
# plt.hist(data1, bins=30, density=True, alpha=0.6, color='g', label='Histograma de data1')
plt.plot(data1Linspace, data1pdf, 'k', linewidth=2, label='Data 1')
plt.plot(data2Linspace, data2pdf, 'k', linewidth=2, label='Data 2')
plt.title('Función de Densidad de Probabilidad (PDF) de data1')
plt.xlabel('Valores')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.show()
```

```python
np.random.seed(6)

population_ages1 = stats.poisson.rvs(loc=18, mu=35, size=150000)
population_ages2 = stats.poisson.rvs(loc=18, mu=10, size=100000)
population_ages = np.concatenate((population_ages1, population_ages2))

minnesota_ages1 = stats.poisson.rvs(loc=18, mu=30, size=30)
minnesota_ages2 = stats.poisson.rvs(loc=18, mu=10, size=20)
minnesota_ages = np.concatenate((minnesota_ages1, minnesota_ages2))

print( population_ages.mean() )
print( minnesota_ages.mean() )
```

```python
stats.ttest_1samp(a = minnesota_ages, popmean = population_ages.mean())
```

```python
sigma = minnesota_ages.std()/math.sqrt(50)  # Sample stdev/sample size

stats.t.interval(0.95,                        # Confidence level
                 df = 49,                     # Degrees of freedom
                 loc = minnesota_ages.mean(), # Sample mean
                 scale= sigma) 
```

```python
np.random.seed(12)
wisconsin_ages1 = stats.poisson.rvs(loc=18, mu=33, size=30)
wisconsin_ages2 = stats.poisson.rvs(loc=18, mu=13, size=20)
wisconsin_ages = np.concatenate((wisconsin_ages1, wisconsin_ages2))

print( wisconsin_ages.mean() )
```

```python
stats.ttest_ind(a= minnesota_ages,
                b= wisconsin_ages,
                equal_var=False) 
```

```python
np.random.seed(11)

before= stats.norm.rvs(scale=30, loc=250, size=100)

after = before + stats.norm.rvs(scale=5, loc=-1.25, size=100)

weight_df = pd.DataFrame({"weight_before":before,
                          "weight_after":after,
                          "weight_change":after-before})

weight_df.describe()   
```

```python
stats.ttest_rel(a = before,
                b = after)
```
